Convert name of all soldiers into capital case

```python
def caps(k):
    ans = k.upper() #property of string and not series
    return ans
df['name'] = df['name'].apply(caps)
df
```

```python
df['date_of_birth']= pd.to_datetime(df['date_of_birth'])
df
```

Convert DoB to Age

```python
import datetime as dt
today=dt.datetime.now()
today
```

```python
def dob2age(dob):
    age= (today-dob).days/365
    age= round(age)
    return age
```

```python
df["Age"] = df['date_of_birth'].apply(dob2age)
df
```
