## Filtering

```python
data = {"Name":["Liam","Olivia","Noah","Emma","Oliver","Ava"],
        "Gender":["M","M","F","M","M","F"],
        "City":["Delhi","Mumbai","Delhi","Bangalore","Bangalore","Chennai"],
       "Age":[58,23,24,31,42,28]}
df = pd.DataFrame(data)
df
```

```python
df['Age']>30
```

```python
df[[True,False,True,False,True,True]]
```

#### Numeric filters

```python
df[df['Age']>30]
```

```python
df[df['Age']>=30]
```

```python
df[df['Age']<30]
```

#### Categorical filter

```python
df[df['Gender']=='M']
```

```python
df[df['Gender']!='M']
```

```python
df[df['Gender']=='M'][df['Age']<30]
```

```python
df[df.Gender=='M'][df.Age<30]
```

```python
# AND condition
df[ (df.Gender=="M") & (df.Age<30) ]
```

```python
# OR condition
df[ (df.Gender=="F") | (df.Age<30) ]
```

```python
data = {"Name":["Liam","Eva","Ava","Noah","Oliver","Bhawana"],
       "Salary":[100,500,800,850,950,980]}
df2 = pd.DataFrame(data)
df2
```

```python
df[df2.Salary>500][df.Age>30]
```

```python
df[df.Gender=='M']
```

```python
df[~(df.Gender=='M')]
```

```python
df[ (df2.Salary>500) & (df.Age>30) ]
```

```python
df2[ (df2.Salary>500) & (df.Age>30) ]
```
